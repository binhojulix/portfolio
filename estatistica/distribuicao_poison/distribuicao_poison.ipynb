```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
k = np.arange(0, 17)
print(k)
```

```python
pmf = poisson.pmf(k, mu=7)
pmf = np.round(pmf, 5)
print(pmf)
```

```python
for val, prob in zip(k,pmf):
    print(f"k-value {val} probabilidade = {prob}")
```

```python
plt.plot(k, pmf, marker='o')
plt.xlabel('k')
plt.ylabel('Probability')

plt.show()
```

```python
cdf = poisson.cdf(k, mu=7)
cdf = np.round(cdf, 3)

print(cdf)
```

```python
for val, prob in zip(k,cdf):
    print(f"k-value {val} has probability = {prob}")
```

```python
plt.plot(k, cdf, marker='o')
plt.xlabel('k')
plt.ylabel('probabilidade acumulada')

plt.show()
```
